```python
import requests
import pandas as pd
import json
```

```python
#exemple ville de Bayonne
city1="Bayonne"
country="France"
r=requests.get(f"https://nominatim.openstreetmap.org/search?format=json&q={city1},{country}")

json_data=r.json()
json_data
```

```python
#exemple ville bayonne pour comprendre la data:
lat_bayonne = json_data[0]['lat']
lon_bayonne = json_data[0]['lon']
print(lat_bayonne)
print(lon_bayonne)
```

```python
# reprenons le travail sur l'ensemble des villes
result=[]
for i in city:
    r=requests.get(f"https://nominatim.openstreetmap.org/search?format=json&q={i},{country}")
    result.append(r.json())
```

```python
output_file="data.json"
with open(output_file, "w") as file:
            json.dump(result, file, indent=4)
```

```python
#exemple de la ville La Rochelle (météo des 8 prochains jours)
lat= 46.1591126
lon= -1.1520434
API_key='88af90c2b71147be4761b7296e4177b6'
r=requests.get(f'https://api.openweathermap.org/data/3.0/onecall?lat={lat}&lon={lon}&exclude=current,minutely,hourly,&appid={API_key}')
json_data=r.json()
json_data
```

```python
#daily.pop Probability of precipitation. The values of the parameter vary between 0 and 1, where 0 is equal to 0%, 1 is equal to 100%
pop_list_la_rochelle=[]
for i in range(8):
    pop_list_la_rochelle.append(json_data['daily'][i]['pop'])
pop_list_la_rochelle
```

```python
#daily.rain (where available) Precipitation volume, mm. Please note that only mm as units of measurement are available for this parameter
rain_list_la_rochelle=[]
for i in range(8):
    if json_data['daily'][i]['pop']!=0:
        rain_list_la_rochelle.append(json_data['daily'][i]['rain'])
    else:
        rain_list_la_rochelle.append(0)
rain_list_la_rochelle
```

```python
# cherchons la météo pour l'ensemble des villes
result=[]
for i,j in zip(df["latitude"],df["longitude"]):
    r=requests.get(f'https://api.openweathermap.org/data/3.0/onecall?lat={i}&lon={j}&exclude=current,minutely,hourly,&appid={API_key}')
    result.append(r.json())

result
```
